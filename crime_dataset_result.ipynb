```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import pandas as pd
```

```python
# crime dataset
```

```python
path='/home/mifeng/FFML-main/ffml/crime/'
```

```python
file = 'eo_30_validation.csv'
```

```python
fullpath = path+file
df=pd.read_csv(fullpath, sep=',',header=None)
meanst = df.values[0]
sdt = df.values[1]
```

```python
fig, ax = plt.subplots()
clrs = sns.color_palette("husl", 5)
plt.ylabel('eo value')
plt.xlabel('task')
plt.title('adult 30% validation')
with sns.axes_style("darkgrid"):
    epochs = list(range(len(meanst)))
    ax.plot(epochs, meanst, label='ffml', c=clrs[0])
    ax.fill_between(epochs, meanst-sdt, meanst+sdt ,alpha=0.3, facecolor=clrs[0])
    ax.legend()
```
